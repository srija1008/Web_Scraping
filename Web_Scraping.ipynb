# Web Scraping

```python
# import Libraries
import numpy as np
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup

import time

from IPython.core.interactiveshell import InteractiveShell
```

```python
URL='https://www.value.today/world/world-top-500-companies'
```

```python
website=requests.get(URL)
```

```python
website.status_code  # A status code 200 means we are allowed to scrape the data
```

```python
websitecontent=website.text
```

```python
websoup=BeautifulSoup(websitecontent)
```

```python
com_names=[]
worldranks=[]
market_cap=[]
headq_of_companies=[]
sectors=[]
annual_rev=[]
annual_net_inc=[]
tot_assets=[]
total_emp=[]
employ_count=[]
ceos=[]

for x in websoup.find_all('div',attrs={'class':'row well views-row'}):
    com_name=x.find('div',attrs={'class':'views-field views-field-title col-sm-12 clearfix'})
    if com_name is None:
        com_names.append(np.NaN)
    else:
        com_names.append(com_name.text)

    
    worldrank=x.find('div',attrs={'class':'views-field views-field-field-world-rank-jan-2020 clearfix col-sm-12'})
    if worldrank is None:
        worldranks.append(np.NaN)
    else:
        worldranks.append(worldrank.text)
    
    mar_cap=x.find('div',attrs={'class':'views-field views-field-field-market-value-jan-2020 clearfix col-sm-12'})
    if mar_cap is None:
        market_cap.append(np.NaN)
    else:
        market_cap.append(mar_cap.text)
    
    headq_of_com=x.find('div',attrs={'class':'views-field views-field-field-headquarters-of-company clearfix col-sm-12'})
    if headq_of_com is None:
        headq_of_companies.append(np.NaN)
    else:
        headq_of_companies.append(headq_of_com.text)
    
    sector=x.find('div',attrs={'class':'views-field views-field-field-company-category-primary clearfix col-sm-12'})
    if sector is None:
        sectors.append(np.NaN)
    else:
        sectors.append(sector.text)
    
    ann_rev=x.find('div',attrs={'class':'views-field views-field-field-annual-revenue clearfix col-sm-12'})
    if ann_rev is None:
        annual_rev.append(np.NaN)
    else:
        annual_rev.append(ann_rev.text)
    
    ann_net_inc=x.find('div',attrs={'class':'views-field views-field-field-annual-net-income-lc clearfix col-sm-12'})
    if ann_net_inc is None:
        annual_net_inc.append(np.NaN)
    else:
        annual_net_inc.append(ann_net_inc.text)
    
    tot_ass=x.find('div',attrs={'class':'views-field views-field-field-total-assets clearfix col-sm-12'})
    if tot_ass is None:
        tot_assets.append(np.NaN)
    else:
        tot_assets.append(tot_ass.text)
    
    tot_emp=x.find('div',attrs={'class':'views-field views-field-field-employee-count clearfix col-sm-12'})
    if tot_emp is None:
        total_emp.append(np.NaN)
    else:
        total_emp.append(tot_emp.text)
    
    emp_count=x.find('div',attrs={'class':'views-field views-field-field-employee-count-as-on-date clearfix col-sm-12'})
    if emp_count is None:
        employ_count.append(np.NaN)
    else:
        employ_count.append(emp_count.text)
    
    ceo=x.find('div',attrs={'class':'views-field views-field-field-ceo clearfix col-sm-12'})
    if ceo is None:
        ceos.append(np.NaN)
    else:
        ceos.append(ceo.text)
```

```python
company_df=pd.DataFrame({'Company_Names':com_names,'World_Ranks':worldranks,
                         'Market_Cap':market_cap,'Headquarter_of_Company':headq_of_companies,
                         'Sector':sectors,'Annual_Rev':annual_rev,'Annual_Net_Inc':annual_net_inc,
                         'Total_Assets':tot_assets,'Total_Employees':total_emp,'Employee_Count_as_on_Date':employ_count,'CEO':ceos})
```
